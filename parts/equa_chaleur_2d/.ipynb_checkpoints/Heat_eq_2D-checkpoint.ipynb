On s'intéresse dans ce notebook au problème de la chaleur en deux dimensions,
$$
\frac{\partial u}{\partial t} = D\left(\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}\right) 
$$
sur un domaine $\Omega = [-L,L]\times[-L,L]$.

Il faut discrétiser les opérateurs différentiels. Sur une grille $(x_i, y_j)$, en temps discrétisé $(t_n)$, en posant $u_{i,j}^n = u(x_i,y_j,t_n)$, on fait
$$
\frac{\partial u}{\partial t} \approx \frac{u_{i,j}^{n+1}-u_{i,j}^{n}}{\Delta t}
$$
et le Laplacien se discrétise via
$$
\Delta u(x_i,y_j,t_n) \approx \frac{u_{i+1,j}^{n+1}+u_{i,j+1}^{n+1}+u_{i-1,j}^{n+1}+u_{i,j-1}^{n+1}-4u_{i,j}^{n+1}}{h^2}
$$

On obtient donc le système d'équations, avec $r=\frac{D\Delta t}{h^2}$:
$$
(1+4r)u_{i,j}^{n+1} - ru_{i+1,j}^{n+1} - ru_{i,j+1}^{n+1} -ru_{i-1,j}^{n+1}-ru_{i,j-1}^{n+1}
=
u_{i,j}^n
$$

Mais il n'a pas trop de sens au bord du domaine. Pour cela, on va utiliser nos conditions aux limites

Pour imposer une condition aux limites de type Dirichlet $f:\partial\Omega\rightarrow\mathbb R$, on la discrétise et on répercute en imposant les valeurs $u_{0,j}$, $u_{N-1,j}$, $u_{i,0}$ et $u_{i,N-1}$ dans le système d'équations.

On introduit des points virtuels $u_{-1,j}$, $u_{N,j}$, $u_{i,-1}$ et $u_{i,N}$, de même valeur que leurs cousins. Cela transforme les équations sur les bords en

\begin{aligned}
(1+3r)u_{0,j}^{n+1} - ru_{1,j}^{n+1} - ru_{0,j+1}^{n+1} -ru_{0,j}^{n+1}-ru_{0,j-1}^{n+1}
&= u_{0,j}^n\\
(1+3r)u_{N-1,j}^{n+1} - ru_{N-1,j+1}^{n+1} -ru_{N-2,j}^{n+1}-ru_{N-1,j-1}^{n+1}
&=
u_{N-1,j}^n\\
(1+3r)u_{i,0}^{n+1} - ru_{i+1,0}^{n+1} - ru_{i,1}^{n+1} -ru_{i-1,0}^{n+1}
&= u_{i,0}^n\\
(1+3r)u_{i,N-1}^{n+1} - ru_{i+1,N-1}^{n+1} -ru_{i-1,j}^{n+1}-ru_{i,N-2}^{n+1}
&=
u_{i,N-1}^n
\end{aligned}

On pose
$$
\mathbf U^n = (u_{i,j})_{1\le i,j\le N}
$$
la matrice d'état du système à l'instant $t_n$.

On cherche un tenseur $\mathbf T$ tel que
$$
\mathbf U^{n} = \mathbf T\mathbf U^{n+1}
$$
soit
$$
u_{i,j}^{n} = \sum_{k,l} t^{k,l}_{i,j} u_{k,l}^{n+1}
$$

On prend $t^{k,l}_{i,j} = -r$ si $(k,l)=(i\pm 1, j\pm 1)$, $1+4r$ si $(k,l) = (i,j)$ en dehors des bords, $1+3r$ sur les bords, $1+2r$ aux coins, et $0$ sinon.

Ainsi, notre méthode aux différences finies revient à itérer à chaque instant discrétisé la relation de récurrence tensorielle
$$
\mathbf U^{n+1} = \mathbf T^{-1}\mathbf U^n
$$

# Implémentation en Python

## Discrétisation : Classe `Domain`

La classe suivante permet de représenter le domaine spatial et temporel discrétisé, avec les attributs relatifs aux paramètres de grille (nombre de points, pas, bornes...).

```python
class Domain():
    """
    Définit le domaine spatio-temporel du problème
    """
    def __init__(self, L, T, J = 512, N = 512):
        # Éléments temporels
        self.T = T # temps de la simulation numérique
        self.N = N # Taille des échantillons temporels
        self.times = np.linspace(0,T,N)
        self.dt = T/N
        
        # Éléments spatiaux
        self.L = L # intervalle où simuler
        self.J = J # Taille des échantillons spatiaux
        self.Xs = np.linspace(-L,L,J)
        self.Ys = np.linspace(-L,L,J)
        self.grid = np.meshgrid(self.Xs, self.Ys)
        self.dx = 1/J
```

## Tenseur $\mathbf T$

La fonction suivante calcule l'inverse du tenseur $\mathbf T$ permettant d'itérer la relation de récurrence précédente.

```python
def tenseur(Omega, diffus = 22.8e-6):
    """
    
    diffus : valeur du coefficient de diffusion
    (valeur par défaut : Fer)
    """
    
    J = Omega.J
    tens = np.zeros((J,J,J,J))
    r = diffus*Omega.dt/Omega.dx**2
    indice_domain = [(i,j) for i in range(J) for j in range(J)]
    for k,l in indice_domain:
        for (i,j) in indice_domain:
            if abs(k-i) == 1 and abs(l-j) == 1:
                tens[i][j][k][l] = -r
            elif k==i and l==j:
                a = i in [0,J-1]
                b = j in [0,J-1]
                if a or b:
                    if a and b:
                        tens[i][j][k][l] = 1+2*r
                    else:
                        tens[i][j][k][l] = 1+3*r
                else:
                    tens[i][j][k][l] = 1+4*r
    return np.linalg.tensorinv(tens)
```

## Conditions de Dirichlet

Une fonction qui prend une matrice d'état et impose ses valeurs au bord.

```python
def enforceDirich(Omega,state, f):
    J = Omega.J
    Xs = Omega.Xs
    Ys = Omega.Ys
    for i in range(J):
        state[i][0] = f(Xs[i], Ys[0])
        state[i][J-1] = f(Xs[i], Ys[J-1])
    for j in range(J):
        state[0][j] = f(Xs[0], Ys[j])
        state[J-1][j] = f(Xs[J-1], Ys[j])
```

## Distribution gaussienne

La fonction suivante représente la fonction gaussienne
$$
(x,y)\longmapsto e^{-(x^2+y^2)}
$$

```python
def distrib(x,y, beta = 2, pos=(0,0)):
    v,w = pos
    return np.exp(-beta*((x-v)**2+(y-w)**2))
```

## Fonction `graphe`

Simple fonction qui dessine le profil d'une fonction sur $\Omega$.

```python
def graphe(Omega, U0):
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.grid(True)
    ax.plot_wireframe(*Omega.grid, U0)
```

Un petit test:

```python
Omega = Domain(1,1,16,8)

U0 = 25+10*distrib(*Omega.grid) - 8*distrib(*Omega.grid, beta=5, pos=(0.5,-0.5))

graphe(Omega, U0)
```

## Itération de la relation de récurrence

Il s'agit d'un simple passage dans une boucle en itérant sur les temps discrets, donnés par l'attribut `times` du domaine passé en argument:

```python
def solve(Omega, U0, f=(lambda x,y: 0)):
    times = Omega.times
    tens = tenseur(Omega)
    enforceDirich(Omega, U0, f)
    sol = [U0]
    for t in times:
        state = np.einsum("ijkl,kl->ij", tens, sol[-1])
        enforceDirich(Omega, state, f)
        sol.append(state)
    return sol
```

## Animation

```python
def evol(Omega, U0, f):
    tens = tenseur(Omega)
    states = solve(Omega, U0, f)
    N = Omega.N
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.grid(True)
    ax.plot_wireframe(*Omega.grid, U0)
    zlims = ax.get_zlim()
    
    def animate(i):
        ax.clear()
        ax.set_zlim(zlims)
        data = ax.plot_wireframe(*Omega.grid, states[i])
        return data
    
    anim = animation.FuncAnimation(fig, animate, frames=N,
            interval=1000/25)
    return anim
        
```
